```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from tqdm import tqdm


# Real-world ChatGPT visit data (in millions)
actual_data = np.array([
    152.7, 266, 616, 1000, 1600, 1800, 1800, 
    1600, 1500, 1400, 1500, 1700, 1700, 1600
])

# Assume a constant to convert visit data to adopter estimates
visits_to_adopters = 5  # conversion rate

# Update the actual_data to reflect estimated adopters
estimated_adopters_data = actual_data / visits_to_adopters

# Introduce a retention factor
retention_factor = 0.3  # 50% of those who try ChatGPT continue using it regularly

# Define the differential equations for the adoption model
def adoption_model(t, y, alpha, beta, epsilon, delta, gamma, theta):
    S, T, O, A, R = y
    dSdt = -alpha * S - beta * A * S + theta * A - gamma * S
    dTdt = alpha * S + delta * O - epsilon * T - delta * T 
    dOdt = beta * A * S - delta * O 
    dAdt = retention_factor * (epsilon * T + delta * O + delta * T) - theta * A + 0.4 * A * np.sin(0.59 * t)
    dRdt = gamma * A
    if 11 <= t:
        dAdt -= 0.1 * A * np.cos(0.1*t)
    return [dSdt, dTdt, dOdt, dAdt, dRdt]


# The initial number of people susceptible to using ChatGPT could be the total potential market size
initial_S = 20000  # A large number representing a potential market size
initial_conditions = [20000, 0, 0, estimated_adopters_data[0], 0]  # Susceptible, Tried, Observed, Adopters, Resistant
t_span = (0, len(estimated_adopters_data) - 1)  # Time span from start month to end month

# Time points corresponding to actual data (assuming monthly intervals starting from 0)
t_eval = np.linspace(0, len(estimated_adopters_data) - 1, len(estimated_adopters_data))

# Define the loss functions
def loss_function(solution, estimated_adopters_data):
    # Compute the loss as the sum of absolute differences
    return np.sum(np.abs(solution - estimated_adopters_data))

def MSE_loss_function(solution, actual):
    # Compute the Mean Squared Error
    return np.mean((solution - estimated_adopters_data)**2)

# Define a function to run the simulation and plot the results
def test_adoption(alpha_values, beta_values, estimated_adopters_data):
    rows = len(alpha_values)
    cols = len(beta_values)
    
    plt.figure(figsize=(20, rows * 5)) 
    
    plot_count = 1
    for alpha in tqdm(alpha_values):
        for beta in beta_values:
            
            params = (alpha, beta, 0.008, 0.0045, -0.003, -0.0005)
            sol = solve_ivp(adoption_model, t_span, initial_conditions, args=params, t_eval=t_eval)
            
            plt.subplot(rows, cols, plot_count)
            plt.plot(t_eval, sol.y[3], label=f'Model Prediction; alpha={alpha}, beta={beta}')
            plt.plot(t_eval, estimated_adopters_data, 'o', label='Actual Visits', markersize=3)
            # Calculate loss and MSE
            loss_val = loss_function(sol.y[3], estimated_adopters_data)
            mse_loss_val = MSE_loss_function(sol.y[3], estimated_adopters_data)
            plt.title(f'ChatGPT Adoption (alpha={alpha}, beta={beta})\nLoss: {loss_val:.2f}, MSE Loss: {mse_loss_val:.2f}')
            plt.xlabel('Time (months)')
            plt.ylabel('Number of Adopters (millions)')
            plt.legend()
            plt.grid(True)
            plot_count += 1
    plt.tight_layout()
    plt.show()


alpha_values = [0.08, 0.1]  
beta_values = [0.1, 0.03]   


test_adoption(alpha_values, beta_values, estimated_adopters_data)
```
